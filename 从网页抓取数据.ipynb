# Data scraping from the Web: Reading data from HTML tables (web-scraping).


## 网页抓取数据

```python
import pandas as pd
import numpy as np

theURL = "http://www.pangoo.it/coronavirus/?t=country&r=ITA&data=y#table"
thedata = pd.read_html(theURL)
# print(thedata)
CovidItaly = pd.DataFrame(thedata[0])
# print(CovidItaly)
```

```python
date = CovidItaly["Giorno"].values
newcases = CovidItaly["Terapia intensiva"].values
print(date)
print(newcases)
```

```python
import matplotlib.pyplot as plt

plt.plot(date,newcases, ls="-", lw=2, label="Covid")
plt.title("New Covid cases in Italy (daily)")
plt.show()
```
